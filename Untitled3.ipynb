```python
def extract_data(dataset):
    data = []
    labels = []

    for image_tensor, label in dataset:
        # Flatten the image tensor and convert to numpy array
        flattened_image = image_tensor.numpy().flatten()
        data.append(flattened_image)
        labels.append(label)

    # Convert lists to numpy arrays
    return np.array(data), np.array(labels)
```

```python
class CNN(nn.Module):
    def __init__(self, num_classes=47, use_bn=True, use_dropout=True, activation_fn=lambda x: x):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, 3, 1, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1, 1)
        
        # Conditional batch normalization
        if use_bn:
            self.bn1 = nn.BatchNorm2d(32)
            self.bn2 = nn.BatchNorm2d(64)
        else:
            self.bn1 = nn.Identity()
            self.bn2 = nn.Identity()

    # Conditional dropout
        if use_dropout:
            self.dropout1 = nn.Dropout(0.5)
            self.dropout2 = nn.Dropout(0.5)
        else:
            self.dropout1 = nn.Identity()
            self.dropout2 = nn.Identity()
        
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 7 * 7, 512)
        self.fc2 = nn.Linear(512, num_classes)
        self.activation_fn = relu  # Assign the lambda function directly without calling it

    def forward(self, x):
        x = self.pool(self.bn1(self.activation_fn(self.conv1(x))))  # Use the lambda function here
        x = self.dropout1(x)
        x = self.pool(self.bn2(self.activation_fn(self.conv2(x))))
        x = self.dropout2(x)
        x = x.view(-1, 64 * 7 * 7)
        x = self.activation_fn(self.fc1(x))  # And here
        x = self.fc2(x)
        return x
```

```python
criterion = nn.CrossEntropyLoss()

def train(network, train_loader, optimizer,criterion,scheduler):
    network.train()
    total_loss = 0.0 
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()

        output = network(data)            
        loss = criterion(output,target) # negative log likelihood loss used for classification
        loss.backward()
        optimizer.step()   
        total_loss += loss.item()

    average_loss = total_loss / len(train_loader)
    print(f'Epoch {epoch}, Average Loss: {average_loss:.4f}')
    scheduler.step()  
```

```python
def accuracy(epoch_idx, data_loader, network, set_type=None):
    correct = 0
    total = 0  # Variable to store the total number of samples processed
    with torch.no_grad():
        for data, target in data_loader:
            data, target = data.to(device), target.to(device)
            outputs = network(data)
            _, predicted = torch.max(outputs.data, 1)
            total += target.size(0)  # Increment the total count
            correct += (predicted == target).sum().item()

    if set_type == "train":
        print(f'\nEpoch{epoch_idx}: Train accuracy: {correct}/{total} ({100. * correct / total:.0f}%)')
    elif set_type == "test":
        print(f'\nEpoch{epoch_idx}: Test accuracy: {correct}/{total} ({100. * correct / total:.0f}%)')

    return correct / total
```

```python
def evaluate(data_loader, network, criterion, set_type="Test"):
    network.eval()  # Set the network to evaluation mode
    correct = 0
    total = 0
    total_loss = 0.0

    with torch.no_grad():
        for data, target in data_loader:
            data, target = data.to(device), target.to(device)
            outputs = network(data)
            loss = criterion(outputs, target)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()

    average_loss = total_loss / len(data_loader)
    accuracy = 100. * correct / total

    print(f'\nEpoch: {set_type} Loss: {average_loss:.4f}, Accuracy: {correct}/{total} ({accuracy:.0f}%)')

    return average_loss, accuracy
```

```python
import torch.nn as nn
import torch.nn.functional as F

# define the NN architecture
class Net(nn.Module):
    def __init__(self, num_classes=47, use_bn=True, use_dropout=True, activation_fn=lambda x: x):
        super(Net, self).__init__()
        # number of hidden nodes in each layer (512)
        self.fc1 = nn.Linear(28 * 28, 512)
        self.fc2 = nn.Linear(512, 512)
        self.fc3 = nn.Linear(512, num_classes)

        # Conditional batch normalization
        self.bn1 = nn.BatchNorm1d(512) if use_bn else nn.Identity()
        self.bn2 = nn.BatchNorm1d(512) if use_bn else nn.Identity()

        # Conditional dropout
        self.dropout1 = nn.Dropout(0.5) if use_dropout else nn.Identity()
        self.dropout2 = nn.Dropout(0.5) if use_dropout else nn.Identity()

        self.activation_fn = activation_fn

    def forward(self, x):
        x = x.view(-1, 28 * 28)  # Flatten image input

        x = self.fc1(x)
        x = self.bn1(x)
        x = self.activation_fn(x)
        x = self.dropout1(x)

        x = self.fc2(x)
        x = self.bn2(x)
        x = self.activation_fn(x)
        x = self.dropout2(x)

        x = self.fc3(x)  # Output layer
        return x
```

```python
criterion = nn.CrossEntropyLoss()

def train(network, train_loader, optimizer,criterion,scheduler):
    network.train()
    total_loss = 0.0 
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data, target
        optimizer.zero_grad()

        output = network(data)            
        loss = criterion(output,target) # negative log likelihood loss used for classification
        loss.backward()
        optimizer.step()   
        total_loss += loss.item()

    average_loss = total_loss / len(train_loader)
    print(f'Epoch {epoch}, Average Loss: {average_loss:.4f}')
    scheduler.step()  
```

```python
def accuracy(epoch_idx, data_loader, network, set_type=None):
    correct = 0
    total = 0  # Variable to store the total number of samples processed
    with torch.no_grad():
        for data, target in data_loader:
            #data, target = data.to(device), target.to(device)
            data, target = data, target
            outputs = network(data)
            _, predicted = torch.max(outputs.data, 1)
            total += target.size(0)  # Increment the total count
            correct += (predicted == target).sum().item()

    if set_type == "train":
        print(f'\nEpoch{epoch_idx}: Train accuracy: {correct}/{total} ({100. * correct / total:.0f}%)')
    elif set_type == "test":
        print(f'\nEpoch{epoch_idx}: Test accuracy: {correct}/{total} ({100. * correct / total:.0f}%)')

    return correct / total
```

```python
def evaluate(data_loader, network, criterion, set_type="Test"):
    network.eval()  # Set the network to evaluation mode
    correct = 0
    total = 0
    total_loss = 0.0

    with torch.no_grad():
        for data, target in data_loader:
            data, target = data, target
            outputs = network(data)
            loss = criterion(outputs, target)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()

    average_loss = total_loss / len(data_loader)
    accuracy = 100. * correct / total

    print(f'\nEpoch: {set_type} Loss: {average_loss:.4f}, Accuracy: {correct}/{total} ({accuracy:.0f}%)')

    return average_loss, accuracy
```
